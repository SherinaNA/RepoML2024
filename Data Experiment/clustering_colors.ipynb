```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#muat dataset colors
colors_df = pd.read_csv('seasonal_colors.csv')
```

```python
# Convert RGB values to a 3D array
rgb_array = colors_df[["R", "G", "B"]].values
```

```python
# Define the number of clusters (seasonal colors)
n_clusters = 4
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(rgb_array)
```

```python
# Get the cluster labels
labels = kmeans.labels_
```

```python
# Create a dictionary to map cluster labels to seasonal colors
seasonal_colors = {
    0: "Spring",
    1: "Summer",
    2: "Autumn",
    3: "Winter"
}
```

```python
# Create a new column in the data frame with the seasonal color labels
colors_df["Seasonal Color"] = [seasonal_colors[label] for label in labels]
```

```python
# Visualize the data using a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(rgb_array[:, 0], rgb_array[:, 1], rgb_array[:, 2], c=labels)
ax.set_xlabel("Red")
ax.set_ylabel("Green")
ax.set_zlabel("Blue")
ax.set_title("Seasonal Colors")
plt.show()
```

```python
# Visualize the data using a 2D scatter plot with seasonal color labels
plt.figure()
plt.scatter(rgb_array[:, 0], rgb_array[:, 1], c=labels)
plt.xlabel("Red")
plt.ylabel("Green")
plt.title("Seasonal Colors")
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', label=seasonal_colors[i], markerfacecolor=plt.cm.tab20(i), markersize=12) for i in range(n_clusters)])
plt.show()
```
